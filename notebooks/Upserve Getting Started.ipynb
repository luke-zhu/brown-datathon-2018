```python
# What is the average variance in sales that a company has?
stds = overall[overall['shift_bin'] == 'Late Night'].groupby('store_id').std()
stds
```

```python
means = overall[overall['shift_bin'] == 'Late Night'].groupby('store_id').mean()
```

```python
(means - stds).sort_values('total_sales')['total_sales'].plot.hist(by='total_sales',
                                                                   title='Late night avg sales minus stdev of provided companies')
```

```python
dinner_df = overall[overall['shift_bin'] == 'Dinner'].groupby('date').mean()
lunch_df = overall[overall['shift_bin'] == 'Lunch'].groupby('date').mean()
```

```python
# Over a year
ax2 = dinner_df.plot.line(use_index=True, y='total_sales', label='dinner')
lunch_df.plot.line(use_index=True, y='total_sales', ax=ax2, label='lunch')
```

```python
# OVer a week
ax3 = dinner_df.iloc[0:28].plot.line(use_index=True, y='total_sales', label='dinner')
lunch_df.iloc[0:28].plot.line(use_index=True, y='total_sales', ax=ax3, label='lunch')
```

```python
shifts = list(by_category['shift_bin'].unique())
categories = list(by_category['item_category'].unique())
```

```python
ax4 = None
for shift in shifts:
    if shift != 'Dinner': continue
    for category in categories:
        filtered = by_category[by_category['shift_bin'] == shift][by_category['item_category'] == category]
        filtered = filtered.groupby('date').mean()
        ax4 = filtered.iloc[0:28].plot.line(use_index=True, y='total_sales', ax=ax4, label='{}|{}'.format(shift,category))
```
